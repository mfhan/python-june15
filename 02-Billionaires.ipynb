## 4a) How many male billionaires are there compared to the number of female billionares?

> **TIP:** The last part uses `groupby`, but the count/percent part does not.
> **TIP:** When I say "average," you can pick what kind of average you use.

```python
#create subgroup of men
male = df[df.gender == "male"]
len(male)
```

```python
#create subgroup of women
fem = df[df.gender == "female"]
len(fem)
```

## 4b) What percent is that?

```python
len(male)/len(fem)
ratio = len(male)/len(fem)
ratio = round(ratio,1)
ratio
```

## 4c) Is the average wealth different for a female billionaire as compared to a male billionaire?

```python
malemed = male.networthusbillion.median()
malemed
```

```python
malemean = male.networthusbillion.mean()
malemean
```

```python
femmean = fem.networthusbillion.mean()
femmean
```

```python
femmed = fem.networthusbillion.median()
femmed
```

```python
femmean > malemean
```

```python
femmed > malemed
```
